**Data Cleaning.**    

1)Handle Missing Data: Impute missing values or remove incomplete records.       
2)Normalize Units: Ensure consistency in units across datasets (e.g., percentage of coverage, number of reported cases).   
3)Date Consistency: Format date fields uniformly across tables for easier analysis.  


```python
import pandas as pd
import os
import re

# Your raw DataFrames
datasets = [df1, df2, df3, df4, df5]
cleaned_datasets = []

# Output directory
output_dir = "cleaned_data"
os.makedirs(output_dir, exist_ok=True)

def clean_dataframe(df, i):
    # 1. Make copy
    df_clean = df.copy()

    # 2. Standardize column names (remove spaces, upper case, underscores)
    df_clean.columns = (
        df_clean.columns
        .str.strip()
        .str.upper()
        .str.replace(r'[^A-Z0-9]+', '_', regex=True)
    )

    # 3. Replace empty strings with NaN
    df_clean = df_clean.replace(r'^\s*$', pd.NA, regex=True)

    # 4. Impute missing values
    num_cols = df_clean.select_dtypes(include='number').columns
    for col in num_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

    cat_cols = df_clean.select_dtypes(exclude='number').columns
    for col in cat_cols:
        mode_value = df_clean[col].mode()
        if not mode_value.empty:
            df_clean[col] = df_clean[col].fillna(mode_value[0])

    # 5. Parse date-like columns safely (exclude YEAR)
    date_like = [col for col in df_clean.columns if "DATE" in col]
    for col in date_like:
        try:
            df_clean[col] = pd.to_datetime(df_clean[col])
        except (ValueError, TypeError):
            continue

    # 6. Ensure YEAR stays integer if present
    if "YEAR" in df_clean.columns:
        df_clean["YEAR"] = pd.to_numeric(df_clean["YEAR"], errors="coerce").astype("Int64")

    # 7. Remove duplicates
    df_clean = df_clean.drop_duplicates()

    # 8. Save to CSV
    filename = os.path.join(output_dir, f"cleaned_df{i}.csv")
    df_clean.to_csv(filename, index=False)
    print(f"âœ… Cleaned and saved dataframe {i} as {filename}")

    return df_clean

# Apply cleaning to all datasets
for i, df in enumerate(datasets, start=1):
    cleaned_datasets.append(clean_dataframe(df, i))

# Assign cleaned versions
df1_clean, df2_clean, df3_clean, df4_clean, df5_clean = cleaned_datasets

print("\nðŸŽ‰ All datasets cleaned and saved successfully!")
```
