```python
import pandas as pd
import os

# List all original dataframes
datasets = [df1, df2, df3, df4, df5]
cleaned_datasets = []

# 1. Create output directory
output_dir = "cleaned_data"
os.makedirs(output_dir, exist_ok=True)

# 2. Clean datasets and save to files
for i, df in enumerate(datasets, start=1):
    # Create working copy
    df_clean = df.copy()
    
    # Handle Missing Data
    df_clean = df_clean.replace(r'^\s*$', pd.NA, regex=True)
    
    # Impute numerical columns with median
    num_cols = df_clean.select_dtypes(include='number').columns
    for col in num_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    
    # Impute categorical columns with mode
    cat_cols = df_clean.select_dtypes(exclude='number').columns
    for col in cat_cols:
        mode_value = df_clean[col].mode()
        if not mode_value.empty:
            df_clean[col] = df_clean[col].fillna(mode_value)
    
    # Normalize Units (customize with your specific conversions)
    unit_conversions = {
        # Example conversions (uncomment and modify as needed):
        # 'coverage': (0.01, 'decimal'),
        # 'reported_cases': (1000, 'thousands'),
    }
    
    for col, (factor, new_name) in unit_conversions.items():
        if col in df_clean.columns:
            df_clean[col] = df_clean[col] * factor
            df_clean = df_clean.rename(columns={col: f"{col}_{new_name}"})
    
    # Date Consistency
    date_cols = [col for col in df_clean.columns if 'date' in col.lower()]
    for col in date_cols:
        df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
    
    # Remove duplicates
    df_clean = df_clean.drop_duplicates()
    
    # Save cleaned dataframe
    filename = os.path.join(output_dir, f"cleaned_df{i}.csv")
    df_clean.to_csv(filename, index=False)
    print(f"✅ Cleaned and saved dataframe {i} as {filename}")
    
    # Add to cleaned datasets list
    cleaned_datasets.append(df_clean)

# 3. Assign cleaned versions to variables
df1_clean, df2_clean, df3_clean, df4_clean, df5_clean = cleaned_datasets

print("\n✅ All datasets cleaned and saved successfully!")
```
